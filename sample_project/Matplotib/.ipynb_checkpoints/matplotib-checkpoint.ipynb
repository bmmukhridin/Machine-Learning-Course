## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax  = plt.subplots(figsize=(10,10)) #(width and height)

# 3. plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. save & show (you save the whole figure)
fig.savefig("../sample-plot.png")
```

## Making figures NumPy arrays

We want:
* Line plot 
* Scatter Plot
* Bar PLot 
* Histogram 
* Subplot 

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0,10,100)
x[:10]
```

```python
fig, ax = plt.subplots()
ax.plot(x,x**2);
```

```python
# use same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {
    "Almond butter":10,
    "Peanut butter":8,
    "Cashew butter":12
}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="Mike's Nut butter Store",
      ylabel="Prices ($)");
```
