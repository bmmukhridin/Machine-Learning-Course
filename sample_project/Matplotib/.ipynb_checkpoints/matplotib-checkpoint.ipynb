## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax  = plt.subplots(figsize=(10,10)) #(width and height)

# 3. plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. save & show (you save the whole figure)
fig.savefig("../sample-plot.png")
```

## Two options for sublots

```python
# Subplot option 1 
fig,((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
# Plot to each diffrent axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.random(100));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))

# Plot to each diffrent index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.random(100));
```

## Which one should you use? (pyplot vs matplotlib OO method?)

* When plotting somthing quickly, ok to use the pyplot method.
* When plotting somthing more advanced, use the OO method.

```python
over_50 = heartDisease[heartDisease["age"]>50]
over_50.head()
```

```python
# PyPlot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
# OO method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([45,100])
```

```python
# OO method from scratch

fig, ax=plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Choleterol");

# Add legend 
ax.legend(*scatter.legend_elements(), title="Target");

# Add horizonal line
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
fig, ax=plt.subplots(figsize=(10,6))

# Plot the data
scatter2 = ax.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"]);
# Customize
ax.set(title="MAX heart beat",
      xlabel="Age",
      ylabel="Thalach");
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10))

# Add data to ax0
scatter2 = ax0.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"]);

# customize ax0
ax0.set(title="MAX heart beat",
      xlabel="",
      ylabel="Thalach");

# add legend ax0
ax0.legend(*scatter.legend_elements(), title="Target");

# Add horizonal line ax0
ax0.axhline(over_50["thalach"].mean(), linestyle="--");

# --------------------------------------------------------

# Add data to ax1
scatter3 = ax1.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);

# customize ax1
ax1.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Thalach");

# add legend ax1
ax1.legend(*scatter.legend_elements(), title="Target");

# Add horizonal line ax1
ax1.axhline(over_50["chol"].mean(), linestyle="--");

# Add super title
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
